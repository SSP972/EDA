```python
#checking unique values
for ind,col in enumerate( df.columns):    
    print(ind," ",col ,'-------',
        df[col].unique())
```

```python
def convert_Quantifiable(df):
    for col in df.columns:
        df[col]=df[col].astype(str)
        df[col]=df[col].apply(lambda x : int(x.split("'")[1]))
    return df
df= convert_Quantifiable(df)
```

```python
def null_value_list(df):
    """"Enter the DataFrame and get list of null values columns"""
    null_value_column=[col  for col in df.columns if df[col].isnull().sum()>0 ]
    if  len(null_value_column)==0:
        print("No null values presents")                            #this can be converted to logs
    else:
        print(f"This columns have null values {null_value_column}") #this can be converted to logs
        return null_value_column
null_value_list(df)
```

```python
#checking the distribution of data
def check_distribution_each_feature(df):
    position=1
    plt.figure(figsize=(20,40))
    for col in df.columns:
        plt.subplot(8,4,position)
        sns.countplot(x=df[col])
        plt.xlabel(col)
        position+=1
    plt.show()
    plt.savefig(f'countplot{df.shape}')
check_distribution_each_feature(df)
```

#### We have well balance data for all categorical features are already lable encoded as per given in the documentation (href='https://view.officeapps.live.com/op/view.aspx?src=https%3A%2F%2Farchive.ics.uci.edu%2Fml%2Fmachine-learning-databases%2F00327%2FPhishing%2520Websites%2520Features.docx&wdOrigin=BROWSELINK')

```python
#checking the distribution of data w.r.t. output feature
def check_features_wrt_output(df):
    position=1
    plt.figure(figsize=(20,40))
    for col in df.columns:
        plt.subplot(8,4,position)
        sns.violinplot(data=df,x=df[col],y=df['Result'])
        plt.xlabel(col)
        position+=1
    plt.show()
    plt.savefig(f"violinpot{df.shape}")
check_features_wrt_output(df)
```
